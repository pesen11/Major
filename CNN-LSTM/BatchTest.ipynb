```python
from tensorflow.keras.models import load_model
```

```python
model=load_model('phish_detectorv2.h5')
```

```python
import pandas as pd
```

```python
sites=pd.read_csv('dataset_phishing.csv')
```

```python
sites.info()
```

```python
sites.tail()
```

```python
sites.columns
```

```python
sites=sites[['url','status']]
```

```python
sites.head()
```

```python
label=[]
for i in range(0,len(sites)):
  if sites['status'][i]=='phishing':
    label.append(1)
  else:
    label.append(0)
```

```python
sites['label']=label
```

```python
sites.head()
```

```python
sites.tail()
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing import sequence
```

```python
tokenizer = Tokenizer(num_words=300, char_level=True)
tokenizer.fit_on_texts(sites['url'])
```

```python
data = tokenizer.texts_to_sequences(sites['url'])
```

```python
X=sequence.pad_sequences(data, maxlen=300,padding='post')
y=sites['label']
```

```python
X[0]
```

```python
y.head()
```

```python
first=X[:2000]
firstLabel=y[:2000]

second=X[2000:4000]
secondLabel=y[2000:4000]

third=X[4000:6000]
thirdLabel=y[4000:6000]

fourth=X[6000:8000]
fourthLabel=y[6000:8000]

fifth=X[8000:10000]
fifthLabel=y[8000:10000]

sixth=X[10000:11429]
sixthLabel=y[10000:11429]
```

```python
pred1=model.predict(first)
pred2=model.predict(second)
pred3=model.predict(third)
pred4=model.predict(fourth)
pred5=model.predict(fifth)
```

```python
pred6=model.predict(sixth)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_recall_fscore_support
```

```python
print(confusion_matrix((pred1>0.5),firstLabel))
print(precision_recall_fscore_support((pred1>0.5),firstLabel,average='macro'))
print(classification_report((pred1>0.5),firstLabel))
```

```python
print(confusion_matrix((pred2>0.5),secondLabel))
print(precision_recall_fscore_support((pred2>0.5),secondLabel,average='macro'))
```

```python
print(confusion_matrix((pred3>0.5),thirdLabel))
print(precision_recall_fscore_support((pred3>0.5),thirdLabel,average='macro'))
```

```python
print(confusion_matrix((pred4>0.5),fourthLabel))
print(precision_recall_fscore_support((pred4>0.5),fourthLabel,average='macro'))
```

```python
print(confusion_matrix((pred5>0.5),fifthLabel))
print(precision_recall_fscore_support((pred5>0.5),fifthLabel,average='macro'))
```

```python
print(confusion_matrix((pred6>0.5),sixthLabel))
print(precision_recall_fscore_support((pred6>0.5),sixthLabel,average='macro'))
```
